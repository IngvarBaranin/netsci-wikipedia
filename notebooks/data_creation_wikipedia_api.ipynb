```python
import wikipedia
from wikipedia.exceptions import WikipediaException

import pandas
```

```python
# Initiate graph traversing with some beginning article
source_page = wikipedia.page("Philosophy")

# Initiate graph traversal variables
visited = []
queue = []
visited.append(source_page.title)
queue.append(source_page.title)

# Initiate data gathering variables
# For nodes
pop_count = 0

# For links
sources = []
targets = []

# For node info
node_names = []
node_attributes = []

while queue:
    node = queue.pop(0)
    pop_count += 1
    
    print(pop_count, node)
    
    # Get current node's article for its attributes and links
    try:
        node_page = wikipedia.page(node, auto_suggest=False)
    except WikipediaException as e:
        print(e)
        continue
    
    # Remember node attributes
    node_names.append(node_page.title)
    node_attributes.append(node_page.summary)
    
    for link in node_page.links:
        # For data 
        # Links between nodes have to be gathered multiple times, as they can exist multiple times
        sources.append(node)
        targets.append(link)
        
        # Continue BFS until 250 000 unique nodes
        if link not in visited and pop_count < 225_000:
            # For traversing
            visited.append(link)
            queue.append(link)
```
